### Task 1: Exploring Data Distributions Using Bubble Plots


#### 1. Bubble Plot for Age vs. Frequency of Participation


- Visualize the relationship between respondents’ age and their participation frequency (`SOPartFreq`) using a bubble plot.

- Use the size of the bubbles to represent their job satisfaction (`JobSat`).


```python
import seaborn as sns
```

```python
##Write your code here

df_touse = df[(df['Age'].notna()) & (df['SOPartFreq'].notna()) & (df['JobSat'].notna())]

plt.figure(figsize=(10, 6))
sns.scatterplot(
    data=df,
    x='Age',
    y='SOPartFreq',
    size='JobSat',
    sizes=(100, 1000),
    color='skyblue',
    alpha=0.7,
    legend='brief'
)

# Customizing plot
plt.title('Age vs. Stack Overflow Participation Frequency\n(Bubble size = Job Satisfaction)', fontsize=14)
plt.xlabel('Age Group')
plt.ylabel('SO Participation Frequency')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

```

#### 2. Bubble Plot for Compensation vs. Job Satisfaction


-Visualize the relationship between yearly compensation (`ConvertedCompYearly`) and job satisfaction (`JobSat`).

- Use the size of the bubbles to represent respondents’ age.


```python
##Write your code here



# Plotting
plt.figure(figsize=(10, 6))
sns.scatterplot(
    data=df,
    x='ConvertedCompYearly',
    y='JobSat',
    size='Age',
    sizes=(100, 1000),
    hue='Age',
    palette='coolwarm',
    alpha=0.7,
    legend='brief'
)

# Customize plot
plt.title('Yearly Compensation vs. Job Satisfaction\n(Bubble Size = Age)', fontsize=14)
plt.xlabel('Yearly Compensation (USD)')
plt.ylabel('Job Satisfaction')
plt.grid(True)
plt.tight_layout()

# Show plot
plt.show()
```

### Task 2: Analyzing Relationships Using Bubble Plots


#### 1. Bubble Plot of Technology Preferences by Age

- Visualize the popularity of programming languages respondents have worked with (`LanguageHaveWorkedWith`) across age groups.

- Use bubble size to represent the frequency of each language.



```python
# Step 1: Split the 'LanguageHaveWorkedWith' column
df_expanded = df.assign(Language=df['LanguageHaveWorkedWith'].str.split(';')).explode('Language')

# Step 2: Group by Age and Language, count occurrences
lang_age_counts = df_expanded.groupby(['Age', 'Language']).size().reset_index(name='Count')##Write your code here


plt.figure(figsize=(12, 7))
sns.scatterplot(
    data=lang_age_counts,
    x='Age',
    y='Language',
    size='Count',
    sizes=(100, 1000),
    hue='Count',
    palette='viridis',
    alpha=0.7,
    legend='brief'
)

# Customize the plot
plt.title('Popularity of Programming Languages by Age Group\n(Bubble Size = Frequency)', fontsize=14)
plt.xlabel('Age Group')
plt.ylabel('Programming Language')
plt.grid(True)
plt.tight_layout()

# Show plot
plt.show()
```

#### 2. Bubble Plot for Preferred Databases vs. Job Satisfaction

- Explore the relationship between preferred databases (`DatabaseWantToWorkWith`) and job satisfaction.

- Use bubble size to indicate the number of respondents for each database.


```python
##Write your code here
# Step 1: Split multi-valued column into separate rows
df_exploded = df.assign(Database=df['DatabaseWantToWorkWith'].str.split(';')).explode('Database')

# Step 2: Group by database and calculate average Job Satisfaction + count
db_summary = df_exploded.groupby('Database').agg(
    AvgJobSat=('JobSat', 'mean'),
    RespondentCount=('JobSat', 'count')
).reset_index()

# Step 3: Plot bubble chart
plt.figure(figsize=(10, 6))
sns.scatterplot(
    data=db_summary,
    x='Database',
    y='AvgJobSat',
    size='RespondentCount',
    sizes=(100, 1000),
    hue='RespondentCount',
    palette='coolwarm',
    alpha=0.7,
    legend='brief'
)

# Customize plot
plt.title('Preferred Databases vs. Job Satisfaction\n(Bubble Size = Number of Respondents)', fontsize=14)
plt.xlabel('Database')
plt.ylabel('Average Job Satisfaction')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()

# Show plot
plt.show()
```

### Task 3: Comparing Data Using Bubble Plots


#### 1. Bubble Plot for Compensation Across Developer Roles

- Visualize compensation (`ConvertedCompYearly`) across different developer roles (`DevType`).

- Use bubble size to represent job satisfaction.


```python
##Write your code here
plt.figure(figsize=(12, 6))
sns.scatterplot(
    data=df,
    x='DevType',
    y='ConvertedCompYearly',
    size='JobSat',
    sizes=(100, 1000),
    hue='JobSat',
    palette='viridis',
    alpha=0.7,
    legend='brief'
)

# Customize the plot
plt.title('Compensation Across Developer Roles\n(Bubble Size = Job Satisfaction)', fontsize=14)
plt.xlabel('Developer Role')
plt.ylabel('Yearly Compensation (USD)')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()

# Show plot
plt.show()
```

#### 2. Bubble Plot for Collaboration Tools by Age

- Visualize the relationship between the collaboration tools used (`NEWCollabToolsHaveWorkedWith`) and age groups.

- Use bubble size to represent the frequency of tool usage.


```python
##Write your code here
# Step 1: Split the collaboration tools column
df_exploded = df.assign(Tool=df['NEWCollabToolsHaveWorkedWith'].str.split(';')).explode('Tool')

# Step 2: Group by Age and Tool to count frequency
tool_usage = df_exploded.groupby(['Age', 'Tool']).size().reset_index(name='UsageCount')

# Step 3: Plot bubble chart
plt.figure(figsize=(12, 6))
sns.scatterplot(
    data=tool_usage,
    x='Age',
    y='Tool',
    size='UsageCount',
    sizes=(100, 1000),
    hue='UsageCount',
    palette='coolwarm',
    alpha=0.7,
    legend='brief'
)

# Customize the plot
plt.title('Collaboration Tool Usage Across Age Groups\n(Bubble Size = Frequency)', fontsize=14)
plt.xlabel('Age Group')
plt.ylabel('Collaboration Tool')
plt.grid(True)
plt.tight_layout()

# Show plot
plt.show()
```

### Task 4: Visualizing Technology Trends Using Bubble Plots


#### 1. Bubble Plot for Preferred Web Frameworks vs. Job Satisfaction

- Explore the relationship between preferred web frameworks (`WebframeWantToWorkWith`) and job satisfaction.

- Use bubble size to represent the number of respondents.



```python
##Write your code here
# Step 1: Split multi-select responses
df_exploded = df.assign(Framework=df['WebframeWantToWorkWith'].str.split(';')).explode('Framework')

# Step 2: Group by framework and calculate average job satisfaction + count of respondents
framework_summary = df_exploded.groupby('Framework').agg(
    AvgJobSat=('JobSat', 'mean'),
    RespondentCount=('JobSat', 'count')
).reset_index()

# Step 3: Plot bubble chart
plt.figure(figsize=(15, 10))
sns.scatterplot(
    data=framework_summary,
    x='Framework',
    y='AvgJobSat',
    size='RespondentCount',
    sizes=(100, 1000),
    hue='RespondentCount',
    palette='viridis',
    alpha=0.7,
    legend='brief'
)

# Customize plot
plt.title('Preferred Web Frameworks vs. Job Satisfaction\n(Bubble Size = Number of Respondents)', fontsize=14)
plt.xlabel('Web Framework')
plt.ylabel('Average Job Satisfaction')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()

# Show plot
plt.show()
```

#### 2. Bubble Plot for Admired Technologies Across Countries

- Visualize the distribution of admired technologies (`LanguageAdmired`) across different countries (`Country`).

- Use bubble size to represent the frequency of admiration.



```python
##Write your code here
# Group by 'Country' and 'LanguageAdmired' to aggregate the admiration frequency
df_2 = df.dropna(subset=['LanguageAdmired', 'Region'])
grouped_df = df_2.groupby(['Country', 'LanguageAdmired']).size().reset_index(name='Frequency')

# Set up the plot
plt.figure(figsize=(12, 8))

# Create a scatter plot
sns.scatterplot(data=grouped_df, x='Country', y='LanguageAdmired',size='Frequency', sizes=(20, 200), hue='LanguageAdmired', palette='Set1', legend=None, marker='o')

# Customize plot
plt.title('Admired Technologies Across Different Countries', fontsize=16)
plt.xlabel('Country', fontsize=12)
plt.ylabel('Admired Technology (Language)', fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()

# Show the plot
plt.show()
```
