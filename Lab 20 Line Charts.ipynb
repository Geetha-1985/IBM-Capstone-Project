### Task 1: Trends in Compensation Over Age Groups


##### 1. Line Chart of Median `ConvertedCompYearly` by Age Group


- Track how the median yearly compensation (ConvertedCompYearly) changes across different age groups.

- Use a line chart to visualize these trends.


```python
## Write your code here
median_comp = df.groupby('Age')['ConvertedCompYearly'].median().sort_index()

# Plot line chart
plt.figure(figsize=(10, 6))
plt.plot(median_comp.index, median_comp.values, marker='o', linestyle='-', color='teal')

# Customize plot
plt.title('Median Yearly Compensation by Age Group', fontsize=14)
plt.xlabel('Age Group', fontsize=12)
plt.ylabel('Median Compensation (USD)', fontsize=12)
plt.grid(True)
plt.tight_layout()

# Show plot
plt.show()
```

##### 2. Line Chart of Median `ConvertedCompYearly` for Ages 25 to 45


For a closer look, plot a line chart focusing on the median compensation for respondents between ages 25 and 45.


```python
print(df['Age'].unique())
```

```python
## Write your code here
df_filtered = df[(df['Age'] == "25-34 years old") | (df['Age'] == "35-44 years old") | (df['Age'] == "45-54 years old")]
median_by_age = df_filtered.groupby('Age')['ConvertedCompYearly'].median()

# Plot line chart
plt.figure(figsize=(10, 6))
plt.plot(median_by_age.index, median_by_age.values, marker='o', linestyle='-', color='orange')

# Customize chart
plt.title('Median Yearly Compensation for Ages 25 to 45', fontsize=14)
plt.xlabel('Age', fontsize=12)
plt.ylabel('Median Compensation (USD)', fontsize=12)
plt.grid(True)
plt.tight_layout()

# Show plot
plt.show()
```

### Task 2: Trends in Job Satisfaction by Experience Level



##### 1. Line Chart of Job Satisfaction (`JobSatPoints_6`) by Experience Level



- Use a column that approximates experience level to analyze how job satisfaction changes with experience.

- If needed, substitute an available experience-related column for `Experience`.


```python
## Write your code here
# Convert 'YearsCode' to numeric, handle non-numeric if needed
df['YearsCode'] = pd.to_numeric(df['YearsCode'], errors='coerce')

# Drop any rows with missing or invalid YearsCode
df = df.dropna(subset=['YearsCode'])

# Group by YearsCode and calculate median job satisfaction
median_satisfaction = df.groupby('YearsCode')['JobSatPoints_6'].median()

# Plot line chart
plt.figure(figsize=(10, 6))
plt.plot(median_satisfaction.index, median_satisfaction.values, marker='o', linestyle='-', color='purple')

# Customize chart
plt.title('Job Satisfaction by Experience Level (Years of Coding)', fontsize=14)
plt.xlabel('Years of Coding Experience', fontsize=12)
plt.ylabel('Median Job Satisfaction (JobSatPoints_6)', fontsize=12)
plt.grid(True)
plt.tight_layout()

# Show plot
plt.show()
```

### Task 3: Trends in Job Satisfaction and Compensation by Experience


##### 1.Line Chart of Median ConvertedCompYearly Over Experience Level

- This line chart will track how median compensation (`ConvertedCompYearly`) changes with increasing experience.

- Use a column such as `WorkExp` or another relevant experience-related column.


```python
## Write your code here
df['YearsCodePro'] = pd.to_numeric(df['YearsCodePro'], errors='coerce')

# Drop rows with missing or invalid WorkExp
df = df.dropna(subset=['YearsCodePro'])

# Group by WorkExp and calculate median compensation
median_comp_by_exp = df.groupby('YearsCodePro')['ConvertedCompYearly'].median()

# Plot line chart
plt.figure(figsize=(10, 6))
plt.plot(median_comp_by_exp.index, median_comp_by_exp.values, marker='o', linestyle='-', color='yellow')

# Customize chart
plt.title('Median Yearly Compensation by Experience (WorkExp)', fontsize=14)
plt.xlabel('Years of Experience', fontsize=12)
plt.ylabel('Median Compensation (USD)', fontsize=12)
plt.grid(True)
plt.tight_layout()

# Show plot
plt.show()
```

##### 2.Line Chart of Job Satisfaction (`JobSatPoints_6`) Across Experience Levels

- Create a line chart to explore trends in job satisfaction (`JobSatPoints_6`) based on experience level.

- This chart will provide insight into how satisfaction correlates with experience over time


```python
## Write your code here
df['WorkExp'] = pd.to_numeric(df['WorkExp'], errors='coerce')

# Drop rows with missing or invalid WorkExp
df = df.dropna(subset=['WorkExp'])

# Group by WorkExp and calculate median job satisfaction
median_satisfaction_by_exp = df.groupby('WorkExp')['JobSatPoints_6'].median()

# Plot line chart
plt.figure(figsize=(10, 6))
plt.plot(median_satisfaction_by_exp.index, median_satisfaction_by_exp.values, marker='o', linestyle='-', color='blue')

# Customize chart
plt.title('Job Satisfaction (JobSatPoints_6) Across Experience Level', fontsize=14)
plt.xlabel('Years of Experience (WorkExp)', fontsize=12)
plt.ylabel('Median Job Satisfaction', fontsize=12)
plt.grid(True)
plt.tight_layout()

# Show plot
plt.show()
```
